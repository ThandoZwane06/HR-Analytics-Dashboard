```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame()
df['color'] = ['blue','red','yellow','green']
df['radius'] = [2,4,3,5]

# Calculate diameter
df['diameter']=df['radius']*2

# Display the updated DataFrame
display(df)

# Now, let's plot a clustered bar chart

# Set position of bar on X axis
bar_width = 0.40
r1 = np.arange(len(df['color']))
r2 = [x + bar_width for x in r1]

fig, ax = plt.subplots(figsize=(6, 3))

# Make the plot
ax.bar(r1, df['radius'], color='skyblue', width=bar_width, edgecolor='yellow', label='Radius')
ax.bar(r2, df['diameter'], color='lightcoral', width=bar_width, edgecolor='grey', label='Diameter')

# Add xticks on the middle of the group bars
ax.set_xlabel('Color', fontweight='bold')
ax.set_ylabel('Value', fontweight='bold')
ax.set_xticks([r + bar_width / 2 for r in range(len(df['color']))])
ax.set_xticklabels(df['color'])
ax.set_title('Clustered Bar Chart: Radius vs Diameter by Color')

# Create legend & Show graphic
ax.legend()
plt.tight_layout()
plt.show()

df_new = df['radius'].sum()
print(df_new)
```

```python

#import the right libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#create a dummy dataset
Sales_data = {
    "Year":[2020,2021,2022,2023,2024,2025,2026],
    "Apple":[100,120,90,200,160,130,110],
    "Samsung":[70,110,140,170,160,140,190],
}

#create a dataframe
df = pd.DataFrame(Sales_data)

#set the position of the bar in the chart
bar_width = 0.35
x1 = np.arange(len(df['Year']))
x2 = [i + bar_width for i in x1]

#create the "frame" and "canvas" first
fig, ax = plt.subplots(figsize=(10,6))

#make the plot
ax.bar(x1,df['Apple'],color='skyblue', width=bar_width, edgecolor='grey', label='Apple' )
ax.bar(x2,df['Samsung'],color='lightcoral', width=bar_width, edgecolor='grey', label='Samsung')

#Mean
mean_apple = df['Apple'].mean()
mean_samsung = df['Samsung'].mean()
combine_mean = (mean_apple + mean_samsung)/2

#add the mean line
ax.axhline(y= combine_mean, color='grey', linestyle='--', label='Average')

#add combine_mean on the chart
mean_text = str(round(combine_mean,0)) #round to nearest whole number

ax.text(x= 0, y= combine_mean, s= mean_text, color = 'black', fontweight= 'bold', va = 'bottom', ha = 'right')


#data labels
ax.set_xlabel('Year', fontweight = 'bold')
ax.set_ylabel('Sales(in Millions)', fontweight = 'bold')
ax.set_title('Sales Comparison between Apple and Samsung', fontweight = 'bold')

#xticks
ax.set_xticks([i + bar_width/2 for i in range(len(df['Year']))])
ax.set_xticklabels(df['Year'])

#legend
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame()
df['name']=['a','b','c','d']
df['salary']=[10,20,30,50]
df['bonus']=[0.2,0.5,0.3,0.8]
df['bonus_salary']=df.eval('salary*bonus')
add = 10
df['bonus_salary_new']=df.eval('salary + @add')

# To see the result of the query, explicitly display it
display(df.query('salary>20')[['name','salary']])

# Group by name and sum salary
summed_salaries = df.groupby('name')['salary'].sum()

# Sort in ascending order
display(summed_salaries.sort_values(ascending=True))

# Sort in descending order
display(summed_salaries.sort_values(ascending=False))

df
```
